```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel("/content/drive/MyDrive/MDP/Data Collection for Recommendation sys (Responses).xlsx")
email = data[['Email Address']]
data
```

```python
data.columns
```

```python
data = data[['10th  Percentage', '12th  Percentage',
       'Batchelors Percentage ', 'MCA CPI (current)',
       'Language (read/ write/ speak)', 'Programming Language', 'FrameWorks','Technologies used on project 1','Technologies used on project 2','Technologies used on project 3']]
data
```

```python
data["10th  Percentage"] = np.where(data["10th  Percentage"]<1,data["10th  Percentage"]*100,data["10th  Percentage"])
data["12th  Percentage"] = np.where(data["12th  Percentage"]<1,data["12th  Percentage"]*100,data["12th  Percentage"])
data['Batchelors Percentage ']=np.where(data["Batchelors Percentage "]<1,data["Batchelors Percentage "]*100,data["Batchelors Percentage "])
```

```python
data['Language (read/ write/ speak)'].value_counts()
```

```python
data['Language (read/ write/ speak)']=data['Language (read/ write/ speak)'].str.replace(" ","")
```

```python
new = data['Language (read/ write/ speak)'].str.split(pat = ",", expand=True).apply(lambda x : x.value_counts(), axis = 1).fillna(0).astype(int)
new1 = data['Programming Language'].str.split(pat = ",", expand=True).apply(lambda x : x.value_counts(), axis = 1).fillna(0).astype(int)
new2 = data['FrameWorks'].str.split(pat = ",", expand=True).apply(lambda x : x.value_counts(), axis = 1).fillna(0).astype(int)
data1=pd.concat([data, new,new1,new2], axis = 1)
```

```python
data1.columns
```

```python
data = data1[['10th  Percentage', '12th  Percentage',
       'Batchelors Percentage ', 'MCA CPI (current)',
       'English', 'Hindi', 'Konkani', 'Maithili', 'Marathi',
       'PASHTO/PERSIAN(DARI)', ' C', ' C++', ' Dart', ' HTML CSS', ' Java',
       ' Java Script', ' Kotlin', ' MYSQL', ' MongoDB ', ' Php', ' R', 'C',
       'Java', 'Python', ' ASP.Net', ' Bootstrap', ' Django', ' Express',
       ' Flask', ' Laravel', 'Bootstrap', 'Express', 'Flutter']]
```
